```python
import pandas as pd
import numpy as np
from transformers import pipeline
import torch
from sentence_transformers import SentenceTransformer
```

```python
df_videos = pd.read_csv('/home/eder/projects/big-data-management/Hackathon_GESTDB/data/database_data/table_video_data.csv')
df_comments = pd.read_csv('/home/eder/projects/big-data-management/Hackathon_GESTDB/data/api_data/comments_data.csv')
print(df_videos.head())
print(df_comments.head())
```

```python
print(df_videos.columns)
print(df_comments.columns)
```

```python
title_embeddings = []
description_embeddings = []
topic_embeddings = []
comment_embeddings = []
```

```python
embedding_dummy = np.zeros((384,))
model = SentenceTransformer('all-MiniLM-L6-v2')
for row in df_videos.itertuples():
    try:
        title_emb = model.encode(row.title_processed)
    except:
        title_emb = embedding_dummy
    try:
        description_emb = model.encode(row.description)
    except:
        description_emb = embedding_dummy
    try:
        topic_emb = model.encode(row.topic)
    except:
        topic_emb = embedding_dummy


    title_embeddings.append([float(x) for x in title_emb.tolist()])
    description_embeddings.append([float(x) for x in description_emb.tolist()])
    topic_embeddings.append([float(x) for x in topic_emb.tolist()])

```

```python
len(topic_embeddings)
```

```python
embedding_dummy = np.zeros((384,))
for row in df_comments.itertuples():
    try:
        comment_emb = model.encode(row.text)
    except:
        comment_emb = embedding_dummy

    comment_embeddings.append([float(x) for x in comment_emb.tolist()])
```
