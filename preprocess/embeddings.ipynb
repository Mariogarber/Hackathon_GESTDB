```python
import pandas as pd
import numpy as np
from transformers import pipeline
import torch

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
embedding = model.encode('The quick brown fox jumps over the lazy dog')
embedding
```

```python
df_videos = pd.read_csv('/home/eder/projects/big-data-management/Hackathon_GESTDB/data/database_data/table_video_data.csv')
df_comments = pd.read_csv('/home/eder/projects/big-data-management/Hackathon_GESTDB/data/api_data/comments_data.csv')
print(df_videos.head())
print(df_comments.head())
```

```python
print(df_videos.columns)
print(df_comments.columns)
```

```python
title_embeddings = []
description_embeddings = []
topic_embeddings = []
comment_embeddings = []
```

```python
embedding_dummy = np.zeros((384,))
for row in df_videos.itertuples():
    try:
        title_emb = model.encode(row.title_processed)
    except:
        title_emb = embedding_dummy
    try:
        description_emb = model.encode(row.description)
    except:
        description_emb = embedding_dummy
    try:
        topic_emb = model.encode(row.topic_processed)
    except:
        topic_emb = embedding_dummy

    title_embeddings.append(title_emb)
    description_embeddings.append(description_emb)
    topic_embeddings.append(topic_emb)
```

```python
embedding_dummy = np.zeros((384,))
for row in df_comments.itertuples():
    try:
        comment_emb = model.encode(row.text)
    except:
        comment_emb = embedding_dummy

    comment_embeddings.append(comment_emb)
```

```python
title_embeddings[0].shape
```

```python
title_embeddings_array = np.array(title_embeddings)
description_embeddings_array = np.array(description_embeddings)
topic_embeddings_array = np.array(topic_embeddings)
comment_embeddings_array = np.array(comment_embeddings)
```

```python
print(title_embeddings_array.shape)
print(description_embeddings_array.shape)
print(topic_embeddings_array.shape)
print(comment_embeddings_array.shape)
```

```python
df_videos['title_embedding'] = title_embeddings
df_videos['description_embedding'] = description_embeddings
df_videos['topic_embedding'] = topic_embeddings
df_comments['comment_embedding'] = comment_embeddings
```
