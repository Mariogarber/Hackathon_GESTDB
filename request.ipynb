```python
# extract a channel wich url is https://www.youtube.com/@Atodaleche

request = youtube.channels().list(
    part="id",
    forHandle="HISTORY"
)
response = request.execute()
channel_id = response["items"][0]["id"]
print(f"Channel ID: {channel_id}")

#get the topic details
request = youtube.channels().list(
    part="topicDetails",
    id=channel_id
)
response = request.execute()
topics = response["items"][0]["topicDetails"]["topicCategories"]
for topic in topics:
    print(f"Topic: {topic}")

# get all videos from the channel (without date restrictions)
request = youtube.search().list(
    part="id",
    channelId=channel_id,
    maxResults=50,
    type="video"
)
response = request.execute()
video_ids = [item["id"]["videoId"] for item in response["items"]]
print(f"Video IDs: {len(video_ids)}")

# look for every video each category and store it

video_categories = {}
for video_id in video_ids:
    request = youtube.videos().list(
        part="snippet",
        id=video_id
    )
    response = request.execute()
    category_id = response["items"][0]["snippet"]["categoryId"]

    if category_id not in video_categories:
        video_categories[category_id] = []
    video_categories[category_id].append(video_id)

for category_id, vids in video_categories.items():
    print(f"Category ID: {category_id}, Total Videos: {len(vids)}")
```

```python
# get a lot of channels that have videos from categories 25, 27, 28 y 35. I want at least 1000 channels per category.
category_ids = ["25", "27", "28", "35"]
all_channel_ids = set()
for category_id in category_ids:
    print(f"Collecting channels for category {category_id}...")
    next_page_token = None
    channels_for_category = 0
    
    while channels_for_category < 1000:
        request = youtube.search().list(
            part="id",
            type="video",
            videoCategoryId=category_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()
        
        if not response.get("items"):
            print(f"No more videos found for category {category_id}")
            break
            
        for item in response["items"]:
            video_id = item["id"]["videoId"]
            video_request = youtube.videos().list(
                part="snippet",
                id=video_id
            )
            video_response = video_request.execute()
            if video_response.get("items"):
                channel_id = video_response["items"][0]["snippet"]["channelId"]
                if channel_id not in all_channel_ids:
                    all_channel_ids.add(channel_id)
                    channels_for_category += 1
                    if channels_for_category >= 1000:
                        break
        
        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            print(f"Reached end of results for category {category_id}")
            break
    
    print(f"Collected {channels_for_category} channels for category {category_id}")

print(f"Total unique channels collected: {len(all_channel_ids)}")

# extract the topicDetail name for each channel and show the unique topics
topic_names = set()
for channel_id in all_channel_ids:
    request = youtube.channels().list(
        part="topicDetails",
        id=channel_id
    )
    response = request.execute()
    if "items" in response and len(response["items"]) > 0:
        topic_details = response["items"][0].get("topicDetails", {})
        for topic in topic_details.get("topicCategories", []):
            topic_names.add(topic)

print(f"Unique topic IDs collected: {topic_names}")
```

```python
# get the number of views, likes and dislikes for each video
video_stats = {}
for video_id in video_ids:
    request = youtube.videos().list(
        part="snippet,statistics",
        id=video_id
    )
    response = request.execute()
    item = response["items"][0]
    stats = item["statistics"]
    title = item["snippet"]["title"]
    video_stats[video_id] = {
        "title": title,
        "views": stats.get("viewCount", 0),
        "likes": stats.get("likeCount", 0),
        "dislikes": stats.get("dislikeCount", None)
    }
print("Video Statistics:")
for video_id, stats in video_stats.items():
    print(f"Video ID: {video_id}, Title: {stats['title']}, Stats: {stats}")
```
